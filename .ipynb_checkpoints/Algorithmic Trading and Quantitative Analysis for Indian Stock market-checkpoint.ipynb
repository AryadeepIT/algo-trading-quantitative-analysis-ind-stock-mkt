## Calculate % Change of Price Change

```python
LTIM_Close_price = close_price_indstocks.LTIM.copy().to_frame()

# to calculate percentage change
LTIM["Lag1"] = LTIM_Close_price.shift(periods=1)
LTIM["Diff"] = LTIM_Close_price.LTIM.diff(periods=1)
LTIM["% Change"] = LTIM_Close_price.LTIM.pct_change(periods=1).mul(100)


LTIM
```

### Deleting unnecessary columns

```python
del LTIM["Lag1"]
del LTIM["Diff"]

LTIM
```

### Renaming Column names

```python
LTIM.rename(columns={'% Change':'Change'}, inplace=True)
LTIM
```

```python
LTIM_Close_price.LTIM.resample("M").last()   #it will give last day of each month
```

```python
LTIM_Close_price.LTIM.resample("BM").last()   #it will give last business-day of each month except the present 
```

## Simple Return vs Compound return vs Log Return

- Simple return, log return, and compound return are three commonly used measures of investment return, each with their own characteristics and uses.

1. Simple return is the most basic measure of investment return. It is calculated by subtracting the initial value of an investment from its final value, and then dividing by the initial value. The result is a percentage that represents the return earned on the investment over a given period of time. Simple return is easy to calculate and understand, but it does not take into account the effect of compounding.

2. Compound return takes compounding into account, and is calculated by adding together all of the returns earned over a specified period of time, and then calculating the overall return on the initial investment. Compound return is often used to compare the returns of different investments over the same period of time. It is a more accurate measure of investment return than simple return, as it reflects the effect of compounding.

3. Log return is a way of measuring the percentage change in the value of an investment over a given period of time. It is calculated by taking the natural logarithm of the ratio of the ending value of the investment to its starting value. Log return is often used in financial modeling and analysis because it has some nice mathematical properties, such as being additive across time periods. Log return is also useful for analyzing the volatility of an investment, as it can help to normalize the data and make it easier to compare changes in value over time.


⨳ We will see the difference between Simple Return and Log Return and take 'Log Return' as Consideration.

### Creating Basic Dataframe and Do Comparision

```python
df = pd.DataFrame(index=[2020,2021,2022], data=[100,50,95], columns = ["Price"] )
df
```

#### Calculating Simple Return :

```python
simplereturns=df.pct_change().dropna()
simplereturns
```

```python
simplereturns.mean()
```

```python
100*1.2*1.2  # it should be 95 so, mean returns are misleading for simple return
```

#### Calculating Log Return :

```python
logreturns = np.log(df/df.shift(1)).dropna()
logreturns
```

```python
# mean of the log return
logreturns.mean()
```

```python
100*np.exp(logreturns.mean()*2)  # correct return achieved (2022 data)
```
